### 2 畫設計圖  (Class)
- 這步驟在物件導向裡面稱為定義 **"Class"**, 中文稱作 **"類別"** 
- 程式語法說明:
    - 每個Class都需要一個名字, 第一行 class Calculator 定義一個名字叫做Calculator的類別
    - 就像筆電開機需要初始化一樣, Class也需要一個初始化的函數, 第二行 def __init__(self) 就是初始化的函數
    - 第6~24行定義了加減乘除四種功能, 並在函數內實作相對應的演算法. 



```python
class Calculator:
    def __init__(self):
        '''計算機初始化'''
        print('Hello, This is DELL Calculator T-800')

    def add(self, x, y):
        '''執行加法運算。

        這個函數接受兩個數字 x 和 y 作為參數，並返回它們的和。

        參數：
        - x (float/int): 第一個加數。
        - y (float/int): 第二個加數。

        返回值：
        float/int: x 和 y 的和。

        例子：
        >>> calculator = Calculator()
        >>> result = calculator.add(3, 4)
        >>> print(result)
        7
        '''
        result = x + y
        return result
    

    def subtract(self, x, y):
        '''減法'''
        result = x - y
        return result

    def multiply(self, x, y):
        '''乘法'''
        result = x * y
        return result

    def divide(self, x, y):
        '''除法'''
        result = x/y
        return result
```

### 3 根據設計圖生產出計算機實體 (Object)
- 這步驟在物件導向裡面稱為 **"根據Class創建Object"**, 中文稱作 **"實體物件"**
- 程式語法說明:
    - 第二行前半段定義Object的變數名稱 calculator
    - 第二行後半段根據Class生成物件 Calculator(), 並將此物件賦值(assign)給calculator變數
- 想像電腦裡面的狀況是這樣: 
    

```python
calculator1 = Calculator()
```

### 4. 使用計算機功能
- 使用設計圖裡面的實作各種功能
- 程式語法: calculator1.add
    - 物件辨識(calculator1)後面的 **"."** 代表你要使用物件裡面的功能


```python
result_add = calculator1.add(5000,2500)
print('5000+2500=',result_add)
result_add = calculator1.subtract(5000, 2500)
print('5000-2500=',result_add)
```
